```python
# melt를 이용해서 last와 first를 같은 형태의 데이터로 변환한다.
df_first_melt = df_first.melt(id_vars='지역', var_name='기간', value_name='평당분양가격')
df_first_melt
```

```python
# 연도 월 분리하기
date = '2013년12월'
date
```

```python
date.split('년')
```

```python
# 리스트의 인덱싱을 사용해서 연도만 가져옵니다.
# -1이라고하면 마지막 데이터를 가지고 온다.
date.split('년')[-1]
```

```python
date.split('년')[-1].replace('월','')
```

```python
# 추출 함수를 만든다.
def parse_year(date):
    year = date.split('년')[0]
    year = int(year)
    return year

y = parse_year(date)
```

```python
type(y)
```

```python
y
```

```python
def parse_month(date):
    month = date.split('년')[-1].replace('월','')
    month = int(month)
    return month
```

```python
parse_month(date)
```

```python
df_first_melt['연도'] = df_first_melt['기간'].apply(parse_year)
df_first_melt['월'] = df_first_melt['기간'].apply(parse_month)
```

```python
df_first_melt.head(1)
```

```python
df_last.head(1)
```

```python
df_last
```

```python
df_last["분양가격"] = pd.to_numeric(df_last["분양가격(㎡)"],errors='coerce')
```

```python
df_last.columns.to_list()
```

```python
df_last.head(5)
```

```python
df_last["평당분양가격"] = df_last["분양가격"] * 3.3
```

```python
df_last.head(5)
```

```python
df_last.columns.to_list()
```

```python
cols = ['지역명','연도','월','평당분양가격']
```

```python
df_last
```

```python
# copy라고 하면 df_last_prepare의 데이터를 변경하더라도 df_last의 데이터에 영향을 주지 않는다.
df_last_prepare =df_last.loc[df_last['규모구분']=='전체', cols].copy()
```

```python
df_last_prepare.head(1)
```

```python
df_first_prepare = df_first_melt[cols].copy()
```

```python
df_first_prepare
```

```python
# concat을 사용하여 두개의 데이터 df_last_prepare, df_first_prepare를 합친다.

df = pd.concat([df_first_prepare, df_last_prepare])
df.shape
```

```python
df.head()
```

```python
df.tail()
```

```python
df['연도'].value_counts(sort=False)
```

```python
t = pd.pivot_table(df, index='연도',columns='지역명', values='평당분양가격').round()
```

```python
# 위에서 그린 피봇테이블을 히트맵으로 표현해 본다.
import seaborn as sns

# 한글 font 지정
import matplotlib.pyplot as plt

plt.rc('font', family='AppleGothic')
```

```python
plt.figure(figsize=(12,7))
sns.heatmap(t, cmap='Blues', annot=True, fmt='.0f')
# .0f는 float 형태로 표현하는데 소숫점 없이 표현하라는 것이다. fmt는 format이라고 생각하면 됨.
```

```python
# T, transpose() 행과 열을 바꿔준다.
plt.figure(figsize=(12,7))
sns.heatmap(t.T, cmap='Blues', annot=True, fmt='.0f')
```

```python
# groupby로 그려본다.
g = df.groupby(['연도','지역명'])['평당분양가격'].mean().unstack().round()
```

```python
plt.figure(figsize=(12,7))
sns.heatmap(g.T, cmap = 'Greens', annot=True, fmt='.0f')
```

```python
# barplot으로 그려보기
sns.barplot(data = df, x='연도', y='평당분양가격')
```

```python
# pointplot으로 연도별 평당분양가격 그리기
plt.figure(figsize=(12,4))
sns.pointplot(data =df, x='연도', y='평당분양가격', hue='지역명')
# legend가 겹쳐보이는 것을 조절하는 방법
plt.legend(bbox_to_anchor=(1.02, 1), loc=2, borderaxespad=0.)
```
