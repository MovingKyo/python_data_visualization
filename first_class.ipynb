## seaborn으로 시각화하기

```python
import seaborn as sns
```

### 지역별 평당 분양가격

```python
plt.figure(figsize=(10,3))
sns.barplot(data=df_last, x='지역명', y='평당분양가격')
# 기본적으로 평균 가격을 구해준다.
```

```python
sns.barplot(data=df_last, x='연도', y='평당분양가격')
```

```python
# lineplot으로 평당분양가격을 그려본다.
# 사용법은 barplot과 유사하다, 함수명만 다르다.
sns.lineplot(data=df_last, x='연도', y='평당분양가격')
```

```python
#hue 값은 색상을 의미한다.
plt.figure(figsize=(15,5))
sns.lineplot(data=df_last, x='연도', y='평당분양가격', hue='지역명')
```

```python
#subplot
sns.relplot(data=df_last, x='연도', y='평당분양가격', kind='line', col='지역명', col_wrap=4, ci=None)
```

```python
sns.catplot(data=df_last, x='연도', y='평당분양가격', kind='bar', col='지역명', col_wrap=4)
```

```python
# seaborn을 사용하면 subplot을 그리는 것을 간단하게 할 수 있다.
```

### seaborn으로 boxplot, violinplot 그리기

```python
sns.violinplot(data=df_last, x='연도', y='평당분양가격')
```

### lmplot, swarmplot

```python
sns.regplot(data=df_last, x='연도', y='평당분양가격')
#regplot은 scatterplot에 회귀선을 그어주는 기능을 한다.
```

```python
sns.scatterplot(data=df_last, x='연도', y='평당분양가격')
```

```python
sns.lmplot(data=df_last, x='연도', y='평당분양가격', hue='전용면적', col='전용면적', col_wrap=4)
```

```python
plt.figure(figsize=(15,3))
sns.swarmplot(data=df_last, x='연도', y='평당분양가격', hue='전용면적')
```
