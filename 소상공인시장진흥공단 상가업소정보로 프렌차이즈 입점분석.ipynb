```python
# Folium으로 지도 활용하기

import folium
```

```python
# 지도의 중심을 지정하기 위해 위도와 경도의 평균을 구한다.
lat = df_31['위도'].mean()
long = df_31['경도'].mean()
```

```python
m = folium.Map(location=[lat, long])
# 127.039032	37.495593
folium.Marker(
    [37.495593, 127.039032],
    popup='<i>던킨도너츠</i>',
    tooltip="던킨도너츠").add_to(m)
# jupyter notebook에서 한글폰트가 깨져보이면 아래와 같이 html로 저장해서 확인하면 된다.
m.save('index.html')
m
```

## 서울의 배스킨라빈스와 던킨도너츠 매장 분포

```python
# 데이터프레임의 인덱스만 출력한다.
df_31.index
```

```python
# icon=folium.Icon(color=icon_color)로 아이콘 컬러를 변경한다.

m = folium.Map([lat, long])




for i in df_31.index:

    sub_lat = df_31.loc[i,'위도']
    sub_long = df_31.loc[i,'경도']
    
    title = df_31.loc[i,'상호명'] + '_' + df_31.loc[i,'도로명주소']
    
    icon_color = 'blue'
    if df_31.loc[i,'브랜드명'] == '던킨도너츠':    
        icon_color = 'red'
    
    folium.Marker(
        
        [sub_lat, sub_long],
        icon=folium.Icon(color=icon_color),
        popup='f<i>(title)</i>',
        tooltip=title).add_to(m)

    
    
m.save('index.html')
m
```

```python
# markercluster로 표현하기

from folium.plugins import MarkerCluster
```

```python
m = folium.Map([lat, long])
marker_cluster = MarkerCluster().add_to(m)

for i in df_31.index:

    sub_lat = df_31.loc[i,'위도']
    sub_long = df_31.loc[i,'경도']
    
    title = df_31.loc[i,'상호명'] + '_' + df_31.loc[i,'도로명주소']
    
    icon_color = 'blue'
    if df_31.loc[i,'브랜드명'] == '던킨도너츠':    
        icon_color = 'red'
    
    folium.Marker(
        
        [sub_lat, sub_long],
        icon=folium.Icon(color=icon_color),
        popup='f<i>(title)</i>',
        tooltip=title).add_to(marker_cluster)

    
    
m.save('index.html')
m
```
